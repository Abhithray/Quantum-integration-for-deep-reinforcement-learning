```python
import gym
from gym import spaces
from stable_baselines3 import DQN

class AnomalyDetectionEnv(gym.Env):
    def __init__(self, X, y):
        super(AnomalyDetectionEnv, self).__init__()
        self.X = X
        self.y = y
        self.current_step = 0
        self.action_space = spaces.Discrete(2)  # 0: Normal, 1: Anomaly
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(X.shape[1],), dtype=np.float32)

    def reset(self):
        self.current_step = 0
        return self.X[self.current_step]

    def step(self, action):
        reward = 1 if action == self.y[self.current_step] else -1
        self.current_step += 1
        done = self.current_step >= len(self.X)
        obs = self.X[self.current_step] if not done else np.zeros(self.X.shape[1])
        return obs, reward, done, {}

# Initialize the environment
env = AnomalyDetectionEnv(X_train_selected, y_train.to_numpy())

# Train the DRL model
model = DQN('MlpPolicy', env, verbose=1)
model.learn(total_timesteps=10000)

from sklearn.metrics import accuracy_score

# Initialize the environment with test data
test_env = AnomalyDetectionEnv(X_test_selected, y_test.to_numpy())

# Evaluate the model
obs = test_env.reset()
predictions = []
for _ in range(len(X_test_selected)):
    action, _ = model.predict(obs, deterministic=True)
    predictions.append(action)
    obs, _, done, _ = test_env.step(action)
    if done:
        break

# Calculate accuracy
accuracy = accuracy_score(y_test[:len(predictions)], predictions)
print(f"Test Accuracy: {accuracy:.2f}")

```
